```
import re
import string
import codecs

with open("test1.txt","r") as f:

  text = f.read()

story = str(re.sub(r'^https?://*[\r]*â€œ','',text,flags = re.MULTILINE))

char_to_remove = 'â€™!@#$%^&*_{}œ\/()'

story = story.translate(story.maketrans('', '', char_to_remove))

story = story.translate(story.maketrans('\n', ' '))

story = story.replace("\\x9", "")

story = story[150:]

sentences = story.split(".")

sentences = sentences[11:len(sentences)-13]

for sentence in sentences:

  sentence = re.sub('[^A-Za-z0-9]+', '', sentence)

```

```
import gensim
from gensim.models import Word2Vec
import logging

sentences = word2vec.Text8Corpus('test1.txt')

sentences
```

```
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = Word2Vec(min_count=1, size=300, window=5, sg=0, hs=1, cbow_mean=1, sample=0.001, iter=50) # initialize an empty model
model.build_vocab(sentences=sentences)
model.train(sentences=sentences, total_examples=model.corpus_count, epochs=model.iter)
```

```
import numpy as np
vocab_size = len(model.wv.vocab)

embedding_matrix = np.zeros((vocab_size, 300))
for i in range(len(model.wv.vocab)):
    embedding_vector = model.wv[model.wv.index2word[i]]
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical

tokenizer = Tokenizer(num_words=None, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ', char_level=False, oov_token=None, document_count=0)

tokenizer.fit_on_texts(sentences)

def get_sequences_of_tokens(corpus):

  input_sequences = []

  for corpora in corpus:

    encoded_seq = tokenizer.texts_to_sequences([corpora])[0]

    for i in range(1,len(encoded_seq)):
      
      n_gram_seq = encoded_seq[:i+1]

      input_sequences.append(n_gram_seq)

  return input_sequences


encoded_sequences = get_sequences_of_tokens(sentences)

encoded_sequences
```

```
import numpy as np


total_words = len(tokenizer.word_index) + 1

def generate_padded_sequences(input_sequences):
    
    max_sequence_len = max([len(x) for x in input_sequences])
    
    input_sequences = np.array(pad_sequences(input_sequences, maxlen = 15, padding = 'pre'))
    
    predictors, label = input_sequences[:,:-1],input_sequences[:,-1]
    
    label = to_categorical(label, num_classes = total_words)
    
    return predictors, label, max_sequence_len

predictors, label, max_sequence_len = generate_padded_sequences(encoded_sequences)
```

```
import numpy as np
from keras.models import Sequential
from tensorflow.keras import regularizers
from keras.layers import Dense, Dropout, Embedding, LSTM, Bidirectional


#model.add(Embedding(input_dim=embedding_matrix.shape[0], output_dim=embedding_matrix.shape[1],
                      #weights=[embedding_matrix]))
```

```
vocab_size = len(tokenizer.word_index) + 1

model = Sequential()

model.add(Embedding(vocab_size, 100, input_length=14))

model.add(Bidirectional(LSTM(100, return_sequences = True)))

model.add(Dropout(0.2))

model.add(LSTM(100))

model.add(Dense(vocab_size//2, activation='relu', kernel_regularizer=regularizers.l2(0.01)))

model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```
model.fit(predictors,label,epochs=40,verbose=2)
```

```
model.save("story_model.h5")
```
