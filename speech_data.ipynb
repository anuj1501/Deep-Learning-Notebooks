```
import librosa
import os
import json

DATASET_PATH = "dataset"
JSON_PATH = "data.json"
SAMPLES_TO_CONSIDER =  9609# 1 sec. of audio

labels = ["zero","one","two","three","four","five","six","seven","eight","nine"]


num_mfcc=13

n_fft=2048

hop_length=512

"""Extracts MFCCs from music dataset and saves them into a json file.

:param dataset_path (str): Path to dataset
    
:param json_path (str): Path to json file used to save MFCCs

:param num_mfcc (int): Number of coefficients to extract

:param n_fft (int): Interval we consider to apply FFT. Measured in # of samples

:param hop_length (int): Sliding window for FFT. Measured in # of samples

:return:
"""

    # dictionary where we'll store mapping, labels, MFCCs and filenames
data = {
    "mapping": [],
    "labels": [],
    "MFCCs": [],
    "files": []
  }

    # loop through all sub-dirs
for label in labels:
      # save label (i.e., sub-folder name) in the mapping
      
  data["mapping"].append(label)
      
  print("\nProcessing: '{}'".format(label))

  filenames = os.listdir("dataset/" + label)

  dirpath = os.path.join("dataset/" + label)

      # process all audio files in sub-dir and store MFCCs
  for f in filenames:

    file_path = os.path.join(dirpath, f)

        # load audio file and slice it to ensure length consistency among different files
    signal, sample_rate = librosa.load(file_path)

    i = labels.index(label)

        # drop audio files with less than pre-decided number of samples
    if len(signal) >= SAMPLES_TO_CONSIDER:

          # ensure consistency of the length of the signal
      signal = signal[:SAMPLES_TO_CONSIDER]

          # extract MFCCs
      MFCCs = librosa.feature.mfcc(signal, sample_rate, n_mfcc=num_mfcc, n_fft=n_fft,
                                                 hop_length=hop_length)

          # store data for analysed track
      data["MFCCs"].append(MFCCs.T.tolist())
      
      data["labels"].append(i)
      
      data["files"].append(file_path)
      
      print("{}: {}".format(file_path, i))

    # save data in json file
with open(JSON_PATH, "w") as fp:
  json.dump(data, fp, indent=4)





  










```

```
import json
import numpy as np
from sklearn.model_selection import train_test_split


DATA_PATH = "data.json"

MODEL_PATH = "model.h5"

LEARNING_RATE = 0.0001

EPOCHS = 10

BATCH_SIZE = 32

# load the data from .json file

with open(DATA_PATH,"r") as jd:

  data = json.load(jd)

#extract the inputs and labels

X = np.array(data["MFCCs"])

Y = np.array(data["labels"])

#create the train/validation/tests splits

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=42)

X_train,X_val,Y_train,Y_val = train_test_split(X_train,Y_train,test_size=0.1,random_state=42)

#convert inputs from 2-D to 3-D array

#X_train = X_train[...,np.newaxis]

#X_val = X_val[...,np.newaxis]

#X_test = X_test[...,np.newaxis]

X_train = X_train.reshape(X_train.shape[0],X_train.shape[1], X_train.shape[2], 1)

X_val = X_val.reshape(X_val.shape[0],X_val.shape[1], X_val.shape[2], 1)

X_test = X_test.reshape(X_test.shape[0],X_test.shape[1], X_test.shape[2], 1)
```

```
print(X_train.shape)
```

```
from keras import Sequential
from keras.layers import Dense,Conv2D,BatchNormalization,Dropout,MaxPool2D,Flatten
from keras.regularizers import l2
from keras.optimizers import Adam

input_shape = (X_train.shape[1],X_train.shape[2],X_train.shape[3])
#build the CNN model

model = Sequential()

#conv layer1

model.add(Conv2D(32,(3,3),activation="relu",input_shape = input_shape,data_format="channels_last"))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2),strides=(2,2),padding="same"))

#conv layer2

model.add(Conv2D(64,(3,3),activation="relu"))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2),strides=(2,2),padding="same"))

#flatten and feed it into dense layer

model.add(Flatten())
model.add(Dense(128,activation = "relu"))

#softmax layer

model.add(Dense(10,activation="softmax"))

#compile the model

optimiser = Adam(learning_rate = LEARNING_RATE)

model.compile(optimizer = optimiser,loss = "sparse_categorical_crossentropy",metrics = ["accuracy"])

print(model.summary())
```

```
#train the model

history = model.fit(X_train,Y_train,epochs=EPOCHS,batch_size=BATCH_SIZE,validation_data=(X_val,Y_val),verbose=2)
```

```
#evaluate the model
test_error,test_accuracy = model.evaluate(X_test,Y_test)

print("test error is {}".format(test_error))
print("test accuracy is {}".format(test_accuracy))
```

```
#save the model
model.save("sound_model.h5")
```
